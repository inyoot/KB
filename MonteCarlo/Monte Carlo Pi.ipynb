```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0.0, 1.0 + 0.01, 0.01)
y = np.sqrt(1-np.square(x))
```

```python
plt.plot(x, y, '-', lw=2)

plt.axes().set_xlim([0,1.0])
plt.axes().set_ylim([0,1.0])
plt.axes().set_aspect('equal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Circle')
plt.grid(True)
```

```python
n = 10000000

px = np.random.uniform(0.0,1.0,n)
py = np.random.uniform(0.0,1.0,n)

mask1 = py > np.sqrt(1-np.square(px))
plt.plot(px[mask1], py[mask1], 'ro')
print(np.sum(mask1))

mask2 = py <= np.sqrt(1-np.square(px))
plt.plot(px[mask2], py[mask2], 'bo')
print(np.sum(mask2))

plt.plot(x, y, '-', lw=2)
plt.axes().set_xlim([0,1.0])
plt.axes().set_ylim([0,1.0])
plt.axes().set_aspect('equal')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Circle ' + '$\pi = $' + str(4.0*np.sum(mask2)/float(n)))
plt.grid(True)
```
